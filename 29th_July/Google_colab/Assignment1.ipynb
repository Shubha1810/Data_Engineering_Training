# 1.Read it into a pandas DataFrame and print the first 3 rows.

```python
df = pd.read_csv('students.csv')
print("First 3 rows:")
print(df.head(3))
```

 # 2. Calculate Average Scores

Add a new column
Average that calculates the mean score for Math, Science, and
English for each student.
Print the student with the highest average.

```python
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)
print("\nStudent with highest average:")
print(df.loc[df['Average'].idxmax()])
```

 # 3. Filter Students

 Display all students who have scored more than 80 in Science.
 Display all students with an average score below 70.

```python
print("\nStudents with Science score > 80:")
print(df[df['Science'] > 80])

print("\nStudents with average score < 70:")
print(df[df['Average'] < 70])
```

# 4. Sorting

 Sort the students by English score (descending) and print the top 3.

```python
sorted_df = df.sort_values(by='English', ascending=False)
print("\nTop 3 students by English score:")
print(sorted_df.head(3))
```

 # 5. Chart Exercise

 Create a bar chart showing students’ names on the X-axis and their average
 scores on the Y-axis.
 Create a line chart for all students’ scores in Math, Science, and English (use
 df.plot() ).

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(df['Name'], df['Average'], color='skyblue')
plt.xlabel('Student Name')
plt.ylabel('Average Score')
plt.title('Average Scores of Students')
plt.grid(True)
plt.show()


df.set_index('Name')[['Math', 'Science', 'English']].plot(kind='line', marker='o')
plt.title('Subject-wise Scores of Students')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```
